```python
%%writefile Document_Scanner.py
import streamlit as st
import cv2
import pytesseract
import numpy as np
from PIL import Image

st.title('Document Scanner Application')

uploaded_Image = st.file_uploader('Please, Upload an Image to Scan...', type = ['jpg', 'png', 'jpeg'])

#load pytesseract application to perform OCR
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

#for google colab: pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

#define a function to scan an image
def Extract_Text(img):
    text = pytesseract.image_to_string(img)
    return text

if uploaded_Image is not None:
    img = Image.open(uploaded_Image)
    #convert image to array
    image_array = np.array(img)
    #show image on streamlit
    st.image(image_array, caption = 'Uploaded Image...', use_column_width = True)

    with st.spinner('Extracting Text From Your Image...'):
        #call Extract_Text function
        extracted_text = Extract_Text(image_array)
        st.subheader('Text Scanned')
        #st.write(extracted_text)

        text_list = extracted_text.splitlines()
        st.write('organization Name:', text_list[0] + ' ' + text_list[1])
        st.write('ID Card :', text_list[3])
        st.write(text_list[4])
        st.write(text_list[5])
        st.write(text_list[6])
        st.write('Name :',text_list[8])
```
