Task 3:
1. Calculate the average value of a specific column.
2. Find the most frequent value in a categorical column.
3. Group the data by a categorical column and calculate the mean for 
each group

```python
# 1. Calculate the average value of a specific column.
average_value = df['Avg_Price'].mean()
print(average_value)
```

```python
# 2. Find the most frequent value in a categorical column.
most_frequent_value = df['Location'].mode()[0]
print(most_frequent_value)
```

```python
# 3. Group the data by a categorical column and calculate the mean for each group.
grouped_df = df.groupby('Location')['Avg_Price'].mean()
print(grouped_df)
```
