```python
class Customer:
    def __init__(self, name , age ) :
        self.name= name
        self.age=age

    # Getter method for name attribute
    def get_name(self):
        return self.name

    # Setter method for name attribute
    def set_name(self, name):
        self._name = name

    # Getter method for age attribute
    def get_age(self):
        return self.age

    # Setter method for age attribute
    def set_age(self, age):
        self._age = age


class MainBikeRental:
    def __init__(self):
        self._list_of_bikes = []  # List to store bikes
        self._list_of_rentals = []  # List to store rentals

    # Getter method for list_of_bikes attribute
    def get_list_of_bikes(self):
        return self._list_of_bikes

    # Setter method for list_of_bikes attribute
    def set_list_of_bikes(self, bikes):
        self._list_of_bikes = bikes

    # Getter method for list_of_rentals attribute
    def get_list_of_rentals(self):
        return self._list_of_rentals

    # Setter method for list_of_rentals attribute
    def set_list_of_rentals(self, rentals):
        self._list_of_rentals = rentals


    # Method to request bikes
    def request_bike(self, customer, number_of_bikes):
        if customer.get_age() > 6 :  # Checking if customer's age is over 6 years
            if len(self._list_of_bikes) >= number_of_bikes:
                # Deduct number_of_bikes from shop stock
                new_num=len(self._list_of_bikes) - number_of_bikes
                self._list_of_bikes = self._list_of_bikes[number_of_bikes:]
               
                # or 
                #del  self._list_of_bikes[:number_of_bikes]

                print(f"{number_of_bikes} bike(s) rented to {customer.get_name()}.")
            else:
                print("Sorry, wait 10 minutes for the bikes to be available.")
        else:
            print("Customer's age should be over 6 years to rent a bike.")



    # Method to return bikes and calculate bill
    def return_bike(self, rental_time, num_of_bikes, bike):
        total_cost = rental_time * 40 * num_of_bikes  # Calculate total cost
        self._list_of_bikes.extend([bike] * num_of_bikes)  # Add returned bikes to stock
        return total_cost

    # Method to add bikes to the list
    def add_bike(self, bike):
        self._list_of_bikes.append(bike)

    # Method to calculate total cost of bikes in the store
    def total_cost(self):
        return len(self._list_of_bikes) * 40  # Each bike costs $40 for rental

# Example usage:

# Create an instance of MainBikeRental
bike_rental = MainBikeRental()

# Add bikes to the list using the add_bike method
bike_rental.add_bike(bike1)
bike_rental.add_bike(bike2)
bike_rental.add_bike(bike3)


# Create a customer
customer1 = Customer("Alice", 10)

# Request bikes for the customer
bike_rental.request_bike(customer1, 2)  # Renting 2 bikes for the customer
print(bike_rental.get_list_of_bikes())

# Return bikes after rental time and calculate total bill cost
total_bill = bike_rental.return_bike(rental_time=3, num_of_bikes=2, bike=bike1)
print("Total bill:", total_bill)

# Get total cost of bikes in the store 
total_store_cost = bike_rental.total_cost()
print("Total cost of bikes in the store:", total_store_cost)


print(bike_rental.get_list_of_bikes())


```

```python
class Bike :
    def __init__(self, name , price ) :
        self.name= name
        self.price=price

    # Getter method for name attribute
    def get_name(self):
        return self.name

    # Setter method for name attribute
    def set_name(self, name):
        self._name = name

    # Getter method for age attribute
    def get_price(self):
        return self.price

    # Setter method for age attribute
    def set_price(self, price):
        self.price = price




class electricBike (Bike) :

    def __init__(self, name , price , maxspeed) :
        super().__init__( name , price )
        self.maxspeed=maxspeed

    # Getter method for name attribute
    def get_maxspeed(self):
        return self.maxspeed

    # Setter method for name attribute
    def set_maxspeed(self, maxspeed):
        self.maxspeed = maxspeed
   

class NormalBike(Bike):
    def __init__(self, name, price, max_speed):
        super().__init__(name, price)
        self._max_speed = max_speed

    # Getter method for max_speed attribute
    def get_max_speed(self):
        return self._max_speed

    # Setter method for max_speed attribute
    def set_max_speed(self, max_speed):
        self._max_speed = max_speed



# Example usage:

# Creating an electric bike instance
electric_bike = electricBike("Electric Bike 1", 1200, 25)
print("Electric Bike Name:", electric_bike.get_name())
print("Electric Bike Price:", electric_bike.get_price())
print("Electric Bike Max Speed:", electric_bike.get_maxspeed())

# Creating a normal bike instance
normal_bike = NormalBike("Normal Bike 1", 500, 20)
print("\nNormal Bike Name:", normal_bike.get_name())
print("Normal Bike Price:", normal_bike.get_price())
print("Normal Bike Max Speed:", normal_bike.get_max_speed())

```
